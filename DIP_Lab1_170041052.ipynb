```python
#necessary imports
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt 
```

```python
image = cv2.imread('/content/flower.jpg',0)

cv2_imshow(image)
cv2.waitKey(0)
```

```python
h,w=image.shape
```

```python
print(h)
```

```python
def thresholding_function(th, img):

  h,w=image.shape

  for i in range(h):
    for j in range(w):

      if img[i,j] > th:
        img[i,j] = 255
      else:
        img[i,j] = 0
  
  return img
```

```python
adjusted_img = thresholding_function(100, image)
cv2_imshow(adjusted_img)
cv2.waitKey(0)
```

**As we can see the orginal picture had some mid grat level pixels representing the flower, so when we pass this picture to the function with a thresholding value of 100 all the mid level gray pixels get mapped to 255 which is why we see that the flower is dominated by white colors**

```python
#lets load a picture dominated by mid gray level values
image = cv2.imread('/content/image_1.tif',0)

cv2_imshow(image)
cv2.waitKey(0)
```

```python
adjusted_img2 = thresholding_function(220, image)
cv2_imshow(adjusted_img2)
cv2.waitKey(0)
```

**As we can see due to a very high threshold value all the mid gray level pixels get mapped to zero and the image turns very dark**

```python
#let's load a very dark image 
image = cv2.imread('/content/dark_image.tif',0)

cv2_imshow(image)
cv2.waitKey(0)
```

```python
def histogram(img):
  array_count = np.zeros([256])
  h,w=image.shape
  for i in range(h):
    for j in range(w):
      var = img[i,j]
      array_count[var]+=1
  return array_count
```

```python
histo = histogram(image)
```

```python
plt.hist(histo, bins = np.linspace(0, 255, 256)) 
plt.title("histogram") 
plt.show()
```

**As we can see there is a dominance of gray values of the lower range, as we can see that the image is dark**

```python
#let's load a very bright image 
image = cv2.imread('/content/bright.tif',0)

cv2_imshow(image)
cv2.waitKey(0)
```

```python
histo = histogram(image)
```

```python
plt.hist(histo, bins = np.linspace(0, 255, 256)) 
plt.title("histogram") 
plt.show()
```

**As we can see there is a dominance of gray values of the higher range, as we can see that the image is bright**
